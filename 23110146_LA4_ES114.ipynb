# Q1. Plot the Probability Density Functions (PDF) of the following continuous random variables

## a) Uniform PDF

### <center>$f_{X}(x;a,b) = \frac{1}{b-a}$</center>

```python
# @title
# @markdown Enter the a,b pairs separated by space for uniform pdf
ab1 = " 2 11" # @param {type:"string"}
ab2 = "3 15" # @param {type:"string"}
ab3 = "5 12" # @param {type:"string"}

to_plot = [list(map(float,ab1.split())),list(map(float,ab3.split())),list(map(float,ab2.split()))]
def uniform_pdf(a,b,x):
  if x<a or x>b:
    return 0
  return 1/(b-a)
for j in to_plot:
  xs = np.linspace(j[0],j[1],1000)
  ys = [uniform_pdf(j[0],j[1],i) for i in xs]
  plt.plot(xs,ys,label=f'a={j[0]} ; b={j[1]}')
  plt.fill_between(xs, ys, alpha=0.3)

plt.title('Uniform PDF')
plt.xlabel('x')
plt.ylabel('$f_X(x)$')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

## b) Gaussian/Normal PDF

### <center>$f_{X}(x;\sigma^2,\mu) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$</center>

```python
# @title
# @markdown Enter the $\mu$, $\sigma^2$ pairs separated by space for gaussian pdf
musig1 = "0 4" # @param {type:"string"}
musig2 = "12 6" # @param {type:"string"}
musig3 = "5 20" # @param {type:"string"}

def gaussian_pdf(x,mean,variance):
  return np.exp(-(x-mean)**2/(2*variance))/np.sqrt(2*np.pi*variance)
to_plot = [list(map(float,musig1.split())),list(map(float,musig2.split())),list(map(float,musig3.split()))]

xl =[]
xr = []
for j in to_plot:
  lim = int(8*np.sqrt(j[1]))
  xl.append(j[0]-lim)
  xr.append(j[0]+lim)

xs = np.linspace(min(xl),max(xr),1000)
for j in to_plot:
  ys = [gaussian_pdf(i,j[0],j[1]) for i in xs]
  plt.plot(xs,ys,label=f'$\mu$={j[0]} ; $\sigma^2$={j[1]}')
  plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.title('Gaussian PDF')
plt.xlabel('x')
plt.ylabel('$f_X(x)$')

plt.show()
```

## c) Exponential PDF

### <center>$f_{X}(x;\lambda) = \lambda e^{-\lambda x}$</center>

```python
# @title
# @markdown Enter the value of $\lambda$ for exponential pdf
lam1 = "5" # @param {type:"string"}
lam2 = "20" # @param {type:"string"}
lam3 = "10" # @param {type:"string"}

to_plot = [float(lam1),float(lam2),float(lam3)]

def exponential_pdf(x,l):
  return l*np.exp(-l*x)

xl = []
xr = []

for l in to_plot:
  mean = 1/l
  var = 1/(l**2)
  xl.append(max(mean-15*var,0))
  xr.append(mean+15*var)

xs = np.linspace(min(xl),max(xr),1000)
for i in to_plot:
  ys = [exponential_pdf(j,i) for j in xs]
  plt.plot(xs,ys,label=f'$\lambda$={i}')
  plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.title('Exponential PDF')
plt.xlabel('x')
plt.ylabel('$f_X(x)$')
plt.show()
```

## d) Cauchy PDF

### <center>$f_{X}(x;x_{0},\gamma) = \frac{1}{\pi\gamma(1 + (\frac{x-x_{0}}{\gamma})^2)}$</center>

```python
# @title
# @markdown Enter the $x_{0}$ (median), $\gamma$ (scale parameter) pairs separated by space for cauchy pdf
mescl1 = "0 0.5" # @param {type:"string"}
mescl2 = "0 2" # @param {type:"string"}
mescl3 = "-2 1" # @param {type:"string"}
to_plot = [list(map(float,mescl1.split())),list(map(float,mescl2.split())),list(map(float,mescl3.split()))]

def cauchy_pdf(x,median,scale_param):
  return 1/(np.pi*scale_param*(1+ ((x-median)/scale_param)**2))

xl=xr=[]
for i in to_plot:
  median = i[0]
  scale_param = i[1]
  xl.append(median-5*scale_param)
  xr.append(median+5*scale_param)



xs = np.linspace(min(xl),max(xr),1000)
for j in to_plot:
  ys = [cauchy_pdf(x,j[0],j[1]) for x in xs]
  plt.plot(xs,ys,label=f'$x_{0}$={j[0]} ; $\gamma$={j[1]}')
  plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))

plt.title('Cauchy PDF')
plt.xlabel('x')
plt.ylabel('$f_X(x)$')
plt.show()
```

## e) Laplacian PDF

### <center>$f_{X}(x;\mu,b) = \frac{1}{2b}e^{-\frac{|x-\mu|}{b}}$</center>

```python
# @title
# @markdown Enter the $\mu$, $b$ pairs separated by space for laplacian pdf
mub1 = "0 1" # @param {type:"string"}
mub2 = "0 2" # @param {type:"string"}
mub3 = "-5 4" # @param {type:"string"}
to_plot = [list(map(float,mub1.split())),list(map(float,mub2.split())),list(map(float,mub3.split()))]


def laplacian_pdf(x,mean,b):
  return (1/(2*b))*np.exp(-(abs(x-mean))/b)
xl=xr=[]
for j in to_plot:
  xl.append(j[0]-5*j[1])
  xr.append(j[0]+5*j[1])

xs = np.linspace(min(xl),max(xr),1000)
for j in to_plot:
  ys = [laplacian_pdf(x,j[0],j[1]) for x in xs]
  plt.plot(xs,ys,label=f'$\mu$={j[0]} ; $b$={j[1]}')
  plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.title('Laplacian PDF')
plt.xlabel('x')
plt.ylabel('$f_X(x)$')
plt.show()
```

## f) Rayleigh PDF

### <center>$f_{X}(x;\sigma) = \frac{x}{\sigma^2}e^{-\frac{x^2}{2\sigma^2}}$</center>

```python
# @title
# @markdown Enter the $\sigma$ value for rayleigh pdf
sig1 = "0.5" # @param {type:"string"}
sig2 = "1" # @param {type:"string"}
sig3 = "2" # @param {type:"string"}
to_plot = [float(sig1),float(sig2),float(sig3)]

def rayleigh_pdf(x,sigma):
  return (x/sigma**2)*np.exp(-x**2/(2*sigma*sigma))
xr = []
for i in to_plot:
  xr.append(5*i)

xs = np.linspace(0,max(xr),1000)
for sigma in to_plot:
  ys = [rayleigh_pdf(i,sigma) for i in xs]

  plt.plot(xs,ys,label=f'$\sigma = {sigma}$')
  plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.title('Rayleigh PDF')
plt.xlabel('x')
plt.ylabel('$f_X(x)$')
plt.show()
```

# Q2. Plot the Cumulative Distribution Function (CDF) of each continuous random variable above.

## a) Uniform PDF

### <center>$F_{X}(x;a,b) = \frac{x-a}{b-a}$</center>

```python
# @title
# @markdown Enter the a,b pair separated by space for uniform cdf
ab = "2 10" # @param {type:"string"}
a,b = list(map(float,ab.split()))

def uniform_cdf(x,a,b):
  if x<a:
    return 0
  if x>b:
    return 1
  return (x-a)/(b-a)

xs = np.linspace(a-1,b+1,1000)
ys = [uniform_cdf(i,a,b) for i in xs]
ys_pdf = [uniform_pdf(a,b,i) for i in xs]
ys_cumsum = np.cumsum(ys_pdf)
ys_cumsum_cdf = np.array(ys_cumsum/ys_cumsum[-1])
plt.plot(xs,ys,label='original')
plt.plot(xs,ys_cumsum_cdf,'-.',label = 'cumsum')
print('Mean square error =',np.sqrt(np.sum((ys-ys_cumsum_cdf)**2)/len(ys)))
plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.legend()
plt.title('CDF for Uniform PDF')
plt.xlabel('x')
plt.ylabel('$F_X(x)$')
plt.show()
```

## b) Gaussian/Normal PDF

### <center>$f_{X}(x;\sigma^2,\mu) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$</center>
<br>
We will be using np.cumsum() for calculating CDF because, we do not have a closed form expression for the CDF.

```python
# @title
# @markdown Enter the $\sigma^2$, $\mu$ pair separated by space for gaussian cdf
sigmu = "10 2" # @param {type:"string"}
sig,mu = list(map(float,sigmu.split()))

xs = np.linspace(mu-8*np.sqrt(sig),mu+8*np.sqrt(sig),1000)
ys_pdf = [gaussian_pdf(i,mu,sig) for  i in xs]
ys_cumsum = np.cumsum(ys_pdf)
ys_cumsum_cdf = ys_cumsum/ys_cumsum[-1]
plt.plot(xs,ys_cumsum_cdf)
plt.title('CDF for Gaussian PDF')
plt.fill_between(xs, ys_cumsum_cdf, alpha=0.3)
plt.xlabel('x')
plt.ylabel('$F_X(x)$')
plt.show()
```

## c) Exponential CDF

### <center>$F_{X}(x;\lambda) = 1 - e^{-\lambda x}$</center>

```python
# @title
# @markdown Enter the value of $\lambda$ for exponential cdf
lam = "5" # @param {type:"string"}
lam = float(lam)

def exponential_cdf(x,lam):
  return 1-np.exp(-lam*x)
mean = 1/lam
var= 1/lam**2

xs = np.linspace(max(mean-30*var,0),mean+30*var,1000)
ys = np.array([exponential_cdf(i,lam) for i in xs])
ys_pdf = [exponential_pdf(i,lam) for i in xs]
ys_cumsum = np.cumsum(ys_pdf)
ys_cumsum_cdf = np.array(ys_cumsum/ys_cumsum[-1])
plt.plot(xs,ys,label='original')
plt.plot(xs,ys_cumsum_cdf,'-.',label = 'cumsum')
print('Mean square error =',np.sqrt(np.sum((ys-ys_cumsum_cdf)**2)/len(ys)))
plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.title('CDF for Exponential PDF')
plt.xlabel('x')
plt.ylabel('$F_X(x)$')
plt.show()


```

## d) Cauchy CDF

### <center>$F_{X}(x;x_{0},\gamma) = \frac{1}{\pi\gamma(1 + (\frac{x-x_{0}}{\gamma})^2)}$</center>

```python
# @title
# @markdown Enter the $x_{0}$, $\gamma$ pair separated by space for cauchy cdf
mescl = "0 0.5" # @param {type:"string"}
me,scl = list(map(float,mescl.split()))

def cauchy_cdf(i,x0,gam):
  return 0.5 + np.arctan((i-x0)/gam)/np.pi

xs = np.linspace(me-10*scl,me+10*scl,1000)
ys = [cauchy_cdf(i,me,scl) for i in xs]
ys_pdf = [cauchy_pdf(i,me,scl) for i in xs]
ys_cumsum = np.cumsum(ys_pdf)
ys_cumsum_cdf = ys_cumsum/ys_cumsum[-1]
plt.plot(xs,ys,label='original')
plt.plot(xs,ys_cumsum_cdf,'-.',label = 'cumsum')
print('Mean square error =',np.sqrt(np.sum((ys-ys_cumsum_cdf)**2)/len(ys)))
plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.title('CDF for Cauchy PDF')
plt.xlabel('x')
plt.ylabel('$F_X(x)$')
plt.show()
```

## e) Laplacian CDF

### <center>$F_{X}(x;\mu,b) = \begin{cases}
    \frac{1}{2}e^{\frac{x-\mu}{b}},& \text{if } x\leq \mu\\
    1-\frac{1}{2}e^{\frac{\mu-x}{b}},              & x\geq \mu
\end{cases}$</center>

```python
# @title
# @markdown Enter the $\mu$, $b$ pairs separated by space for laplacian pdf
mub1 = "0 1" # @param {type:"string"}
mu,b = list(map(float,mub1.split()))

def laplacian_cdf(i,mu,b):
  if i<mu:
    return 0.5*np.exp((i-mu)/b)
  return 1-0.5*np.exp((mu-i)/b)

xs = np.linspace(mu-5*b,mu+5*b,1000)
ys = [laplacian_cdf(i,mu,b) for i in xs]
ys_pdf = [laplacian_pdf(i,mu,b) for i in xs]
ys_cumsum = np.cumsum(ys_pdf)
ys_cumsum_cdf = ys_cumsum/ys_cumsum[-1]
plt.plot(xs,ys,label='original')
plt.plot(xs,ys_cumsum_cdf,'-.',label = 'cumsum')
print('Mean square error =',np.sqrt(np.sum((ys-ys_cumsum_cdf)**2)/len(ys)))
plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.title('CDF for Laplacian PDF')
plt.xlabel('x')
plt.ylabel('$F_X(x)$')
plt.show()
```

## f) Rayleigh CDF

### <center>$F_{X}(x;\sigma) = 1-e^{-\frac{x^2}{2\sigma^2}}$</center>

```python
# @title
# @markdown Enter the $\sigma$ value for rayleigh cdf
sig = "12.5" # @param {type:"string"}
sig =  float(sig)

def rayleigh_cdf(x,sig):
  return 1-np.exp(-x*x/(2*sig*sig))

xs = np.linspace(0,5*sig,1000)
ys = [rayleigh_cdf(i,sig) for i in xs]
ys_pdf = [rayleigh_pdf(i,sig) for i in xs]

ys_cumsum = np.cumsum(ys_pdf)
ys_cumsum_cdf = ys_cumsum/ys_cumsum[-1]
plt.plot(xs,ys,label='original')
plt.plot(xs,ys_cumsum_cdf,'-.',label = 'cumsum')
print('Mean square error =',np.sqrt(np.sum((ys-ys_cumsum_cdf)**2)/len(ys)))
plt.fill_between(xs, ys, alpha=0.3)
plt.legend(bbox_to_anchor=(1,1))
plt.title('CDF for Rayleigh PDF')
plt.xlabel('x')
plt.ylabel('$F_X(x)$')
plt.show()

```

# Q3. Plot the Cumulative Distribution Function (CDF) of each continuous random variable above.

## a) Uniform CDF

### <center>$f_X(x) = \frac{1}{b-a+1}$<br></center>
###<center>$F_X(x) = \sum_{n=a}^{b} f_X(x)$</center>

```python
# @title
# @markdown minimum value
a = 1 # @param {type:"integer"}
# @markdown maximum value
b = 6 # @param {type:"integer"}


def pmf(i,a,b):
  return 0 if i<a or i>b else 1/(b-a+1)

def plot():
  global a,b
  xs =np.arange(a-2,b+2)
  ys = [pmf(i,a,b) for i in xs]
  ys_cdf = np.cumsum(ys)
  ys = ys_cdf/ys_cdf[-1]
  plt.step(xs,ys)
  plt.xlabel('x')
  plt.ylabel('$F_X(x)$')
  plt.title('Discrete Uniform CDF')
  plt.show()

plot()
```

## b) Bernoulli CDF

### <center>$$ f_X(x)=  \left\{
\begin{array}{ll}
      p & x=1 \\
      1-p & x=0 \\
\end{array}
\right. $$</center>
<br>

### <center>$F_X(x) = \sum_{x=0}^{1} f_X(x)$</center>

```python
# @title
# @markdown probability of success
p = 0.7 # @param {type:"number"}

def cdf(k,p):
  if k < 0:
    return 0
  if k>1:
    return 1
  if k == 0:
    return 1-p
  return 1


def plot():
  global p
  xs = [-3,-2,-1,0,1,2,3,4]
  ys = [cdf(i,p) for i in xs]
  plt.step(xs,ys)
  plt.xlabel('x')
  plt.ylabel('$F_X(x)$')
  plt.title('Bernoulli CDF')
  plt.show()

plot()
```

## c) Binomial CDF


### <center>$f_X(x) = \binom nr p^x q^{n-x}$</center>
<br>

### <center>$F_X(x) = \sum_{x=0}^{n} f_X(x)$</center>



```python
# @title
# @markdown number of trials
n = 20 # @param {type:"integer"}
# @markdown probability of success
p = 0.4 # @param {type:"number"}

f={0:1,1:1}
def factorial(n):
  if n in f:
    return f[n]
  return n*factorial(n-1)

def pmfbin(r,n,p):
  # ncr (p)^r (1-p)^(n-r)
  return np.multiply(np.multiply(np.divide(factorial(n),np.multiply(factorial(n-r),factorial(r))),p**r),(1-p)**(n-r)) # i am using np.multiply, or np.divide to circumvent any error replated to precision or overflow

def plot(n,p):
  xs =np.arange(n+1)
  ys = [pmfbin(i,n,p) for i in xs]
  ys_cumsum = np.cumsum(ys)
  plt.step(xs,ys_cumsum/ys_cumsum[-1])
  plt.xlabel('x')
  plt.ylabel('$F_X(x)$')
  plt.title('Binomial CDF')
  plt.show()

plot(n,p)
plotbin = plot
```

## d) Poisson CDF


### <center>$f_X(x) = \frac{e^{-λ}λ^x}{x!}$</center>
<br>

### <center>$F_X(x) = \sum_{x=0}^{n} f_X(x)$</center>


```python
# @title
# @markdown number of trials
n = 80 # @param {type:"number"}
# @markdown number of successes
l = 35 # @param {type:"number"}

f={0:1,1:1}
def factorial(n):
  if n in f:
    return f[n]
  v= n*factorial(n-1)
  f[n] = v
  return v

def pmfpoi(l,k):
  return np.exp(-l)*(l**k)/factorial(k)
def plot_poisson():
  xs =range(1,n+1)
  ys = [pmfpoi(l,i) for i in xs]
  ys_cumsum = np.cumsum(ys)
  plt.step(xs,ys_cumsum)
  plt.xlabel('x')
  plt.ylabel('$F_X(x)$')
  plt.title('Poisson CDF')

plot_poisson()
plt.show()
```

## e) Geometric CDF


### <center>$f_X(x) = (1-p)^{x-1}p$</center>
<br>

### <center>$F_X(x) = \sum_{x=1}^{n} f_X(x)$</center>


```python
# @title
# @markdown number of trials
n = 30 # @param {type:"integer"}
# @markdown probability of success
p = 0.2 # @param {type:"number"}

def pmf(p,k):
  return np.multiply(np.power(1-p,k-1),p)

def plot():
  global p,n
  xs = np.arange(1,n+1)
  ys = [pmf(p,i) for i in xs]
  ys_cumsum = np.cumsum(ys)
  plt.step(xs,ys_cumsum)
  plt.xlabel('x')
  plt.ylabel('$F_X(x)$')
  plt.title('Geometric CDF')
  plt.show()

plot()
```
